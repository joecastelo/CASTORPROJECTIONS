# Step one, create a class to read and display info from the datafile hdr

Follow the lines, it's all made with pandas


```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
class Readinterfilefromgate:
    def __init__(self,filepath):
        df = pd.read_csv(filepath, sep = ":=", engine = 'python').rename(columns = {'Unnamed: 1' : 'Values'})
        self.showme  = df
        def find(name1):
            return df[df['!INTERFILE '].str.contains(name1)]['Values'].reset_index()['Values']
        self.data = find('data description')[0]
        #THIS PART HERE REQUIRES TO THE FILE IS WRITTEN AS ../OUTPUT/FILE1.SIN
        bp = list(find('name of data file')[0].split('/')[2])
        bp.remove('1')
        self.binarypathfile = ''.join(bp)
        self.bptouse = ''.join(bp).split('.')[0]
        bpext = ''.join(bp).split('.')[1]
        self.bpext = '.{}'.format(bpext)
        #######################################################################
        totalnumberofimages = find('total number of images')
        self.totalnumberofimages = int(totalnumberofimages)
        byteorder = find('byte order')[0]
        if byteorder.upper() == ' LITTLEENDIAN':
            byteorder = '<'
        elif byteorder.upper() == ' BIGENDIAN' :
            byteorder = '>'
        self.byteorder = byteorder
        numberofEW = find('number of energy windows')
        EW = find('energy window')[find('energy window') != find('number of images/energy window').min()]
        self.EnergyWindows = []
        for item in (EW):
            self.EnergyWindows.append(item)
        self.directionofrotation = find('direction of rotation')[0][1:]
        self.startangle = float(find('start angle')[0])
        self.extentofrotation = float(find('extent of rotation')[0])
        bytesperpixel = find('number of bytes per pixel')[0]
        self.bytesperpixel = bytesperpixel[0]
        numberformat = find('number format')[0][1]
        self.numberformat = find('number format')[0]
        self.dtype = []
        self.dtype = str(''.join([byteorder, numberformat, bytesperpixel[1]]))
        self.duration = float(find('study duration')[0])
        self.timeprojection = float(find('time per projection')[0])
        self.acqmode = find('acquisition mode')[0]
        self.orbit = find('orbit')[0]
        self.pixelsizeX = float(find('scaling factor ')[0])
        self.pixelsizeY = float(find('scaling factor ')[1])
        self.matrixsizeX = int(find('matrix size ')[0])
        self.matrixsizeY = int(find('matrix size ')[1])
        self.CamtoCOR = abs(float(find(';head x translation')))*10.
    def oneEW(self, splits):
        namelist = []
        input_image_data = []
        input_total = []
        for i in range(1,splits+1):
            namelist.append('{}{}{}'.format(self.bptouse,i,self.bpext))
        for i in range(0,splits):
            input_filename = namelist[i] 
            dtype = self.dtype # little-endian unsigned integer (2 bytes per pixel)
            x , y = self.matrixsizeX , self.matrixsizeY
        # Reading.
            fid = open(input_filename, 'rb')
            data = np.fromfile(fid, self.dtype)
        #Calculate Dimensions from the file
            z = int(data.shape[0]/(x*y))
            imgdata = data.reshape((z,x,y))
        #Separating Windows
        #Every Window should have the same number of projections
        #dim = numberoftotalimages/NumberofENergywindows
        #file.EnergyWindows is a list of strings
        #file.numberoftotalimages is float
            input_total.append(imgdata[:,:,:])

            fid.close()

        total = 0

        
        for i in range(0,splits):
            total += input_total[i]


        return total   
    def tripleEW(self, splits):
        #almost sure this works with dual energy windows # gotta test ...
        namelist = []
        input_image_data = []
        input_image_data_low = []
        input_image_data_high = []
        input_total = []
        for i in range(1,splits+1):
            namelist.append('{}{}{}'.format(self.bptouse,i,self.bpext))
        for i in range(0,splits):
            input_filename = namelist[i] 
            dtype = self.dtype # little-endian unsigned integer (2 bytes per pixel)
            x , y = self.matrixsizeX , self.matrixsizeY
        # Reading.
            fid = open(input_filename, 'rb')
            data = np.fromfile(fid, self.dtype)
        #Calculate Dimensions from the file
            z = int(data.shape[0]/(x*y))
            imgdata = data.reshape((z,x,y))
        #Separating Windows
        #Every Window should have the same number of projections
        #dim = numberoftotalimages/NumberofENergywindows
        #file.EnergyWindows is a list of strings
        #file.numberoftotalimages is float
            div = int(self.totalnumberofimages/int(self.EnergyWindows[0]))
            if int(self.EnergyWindows[0]) > 2:
        #that would have been the approach if the the files were perfect.
                input_image_data.append(imgdata[0:div,:,:])
                input_total.append(imgdata[:,:,:])
                input_image_data_low.append(imgdata[div:2*div,:,:])
                input_image_data_high.append(imgdata[2*div:z,:,:])
                fid.close()
            else:
                input_image_data.append(imgdata[0:div,:,:])
                input_image_data_low.append(imgdata[div:2*div,:,:])
                fid.close()
        central_img = 0
        low_img = 0
        total = 0
        high_img = 0 
        
        for i in range(0,splits):
            total += input_total[i]
            central_img += input_image_data[i]
            low_img +=  input_image_data_low[i]
            high_img += input_image_data_high[i]

        return total,central_img, low_img, high_img #return all the matrixes        
```

```python
# new_arr = np.random.uniform(-10, 10, size = (3,4,5)).astype(np.int16)
# central = sitk.GetImageFromArray(new_arr)
central = sitk.GetImageFromArray(obj[0])
central.SetSpacing([9.59,9.59,1])
```

```python
writer = sitk.ImageFileWriter()
```

```python
writer.KeepOriginalImageUIDOn()
```

```python


modification_time = time.strftime("%H%M%S")
modification_date = time.strftime("%Y%m%d")
```

```python
direction = central.GetDirection()

```

```python
series_tag_values = [("0008|0031",modification_time), # Series Time
                  ("0008|0021",modification_date), # Series Date
                  ("0008|0008","DERIVED\\SECONDARY"), # Image Type
                  ("0020|000e", "1.2.826.0.1.3680043.2.1125."+modification_date+".1"+modification_time), # Series Instance UID
                  ("0020|0037", '\\'.join(map(str, (direction[0], direction[3], direction[6],# Image Orientation (Patient)
                                                    direction[1],direction[4],direction[7])))),
                  ("0008|103e", "Created-SimpleITK")] 
```

```python
for i in range(central.GetDepth()):
    image_slice = central[:,:,i]
    # Tags shared by the series.
    for tag, value in series_tag_values:
        image_slice.SetMetaData(tag, value)
    # Slice specific tags.
    image_slice.SetMetaData("0008|0012", time.strftime("%Y%m%d")) # Instance Creation Date
    image_slice.SetMetaData("0008|0013", time.strftime("%H%M%S")) # Instance Creation Time
    # Setting the type to CT preserves the slice location.
    image_slice.SetMetaData("0008|0060", "NM")  # set the type to CT so the thickness is carried over
    
    # (0020, 0032) image position patient determines the 3D spacing between slices.
    image_slice.SetMetaData("0020|0032", '\\'.join(map(str,central.TransformIndexToPhysicalPoint((0,0,i))))) # Image Position (Patient)
    image_slice.SetMetaData("0020|0013", str(i)) # Instance Number

    # Write to the output directory and add the extension dcm, to force writing in DICOM format.
    writer.SetFileName(os.path.join('./CENTRAL','compare'+str(i)+'.dcm'))
    writer.Execute(image_slice)
```

```python
data_directory = './CENTRAL'
series_IDs = sitk.ImageSeriesReader.GetGDCMSeriesIDs(data_directory)

if not series_IDs:
    print("ERROR: given directory \""+data_directory+"\" does not contain a DICOM series.")
    sys.quit(1)
series_file_names = sitk.ImageSeriesReader.GetGDCMSeriesFileNames(data_directory, series_IDs[0])

series_reader = sitk.ImageSeriesReader()
series_reader.SetFileNames(series_file_names)

# Configure the reader to load all of the DICOM tags (public+private):
# By default tags are not loaded (saves time).
# By default if tags are loaded, the private tags are not loaded.
# We explicitly configure the reader to load tags, including the
# private ones.
series_reader.LoadPrivateTagsOn()
image3D = series_reader.Execute()
```

## Some input parameters

```python
nbh = 2 #NUMBER OF HEADS (VERY IMPORTANT)
params = [file.timeprojection,int(file.totalnumberofimages/(int(file.EnergyWindows[0]))),file.directionofrotation,file.startangle,int(file.extentofrotation*nbh)] # Order: [a,b,c,d,e]  
#Energy Windows
scat_flag = 1 #parameter for the function
n = 1 #normalization flag

output_filename = "Pythongenerated" # Define the output filename 
params
```

## Step3: Creating header...


```python
#def CASTOR(input_img_data,params,scat_flag):
#    """ Function to convert an interfile into a CASToR datafile
#        Inputs: scat_flag : 1 if scatter correction included (needs upper and lower
#                        scatter window datafiles)
#                        0 if no scatter correction included   """
    
input_data_EM = obj[1] #image data: Z, X, Y format

MatrixSize1,MatrixSize2 = input_data_EM.shape[2] , input_data_EM.shape[1]  # X, Y

#get the projection duration
proj_time = params[0]

#get number of projections
NumberOfProjections = params[1]

#get directions of rotation
DirectionOfRotation = params[2]

#get StartAngle,ExtentOfRotation

StartAngle,ExtentOfRotation = params[3],params[4]

if scat_flag == 1:
    input_data_SCL = obj[2]
    input_data_SCU = obj[3]

    W_EM = float(file.EnergyWindows[3])-float(file.EnergyWindows[2])
        
        
    W_SCL = float(file.EnergyWindows[6])-float(file.EnergyWindows[5])
    W_SCU = float(file.EnergyWindows[9])-float(file.EnergyWindows[8])
        
    input_data_SCAT = ((input_data_SCL/W_SCL) + (input_data_SCU/W_SCU))*(W_EM/2); #generate scatter projection data:
    input_data_SCAT = input_data_SCAT/proj_time #convert to counts/sec
    
# create CASToR header file
data_filename = output_filename+".Cdh"
binary = output_filename + ".Cdf" # file to convert the binaries
fid = open(data_filename,'w') #open text file for writing
fid.write('Data filename: ' + binary ); #write to file 
fid.write('\n'); #new line

number_events = MatrixSize1 * MatrixSize2 * NumberOfProjections #calculate number of events
fid.write('Number of events: '+ str(number_events)) #write number of events to file
fid.write('\n'); #new line

fid.write('Data mode: 1')  #these three header fields are probably ok to be the same for every export
fid.write('\n'); #new line

fid.write('Data type: SPECT')
fid.write('\n') #new line

fid.write('Start time (s): 0')
fid.write('\n') #new line

fid.write('Duration (s): '+ str(int(proj_time*NumberOfProjections))) #the time per projection* number of projections ( Total time)
fid.write('\n') #new line

fid.write('Scanner name: '+'symbiat2he') #need to point to a .geom file in the castor config folder #working geom file in my castor config
fid.write('\n') #new line

fid.write('Number of bins: '+ str(64) +','+str(64)) #the matrix size # arbitrary?
fid.write('\n') #new line

fid.write('Number of projections: '+ str(NumberOfProjections)) #DUAL HEAD number of projections ( Total time)
fid.write('\n') #new line

fid.write('Head rotation direction: '+ str(DirectionOfRotation)) #CW or CCW rotation
fid.write('\n') #new line

fid.write('Projection angles: ')
# if DUAL HEADED, double the extentofRotation
angles = np.linspace(StartAngle,ExtentOfRotation,int(NumberOfProjections)) #generate linearly spaced list of angles between start and end angles
#TODO
for i in angles[:-1]: #loop through all but last one
    fid.write(str(i)+',')
fid.write(str(angles[np.size(angles)-1])) #last one doesn't need a comma
fid.write('\n') #new line

fid.write('Normalization correction flag: ' + str(n)) #%tell Castor we're including normalisation (currently set to 1.0 for all events)
fid.write('\n') #new line

fid.write('Scatter correction flag: '+ str(scat_flag)) #scatter correction flag
fid.write('\n') #new line

fid.write('Global distance camera surface to COR: '+ str(file.CamtoCOR)) #globalCORdistancefromcamera
fid.write('\n') #new line

fid.write('Calibration Factor: 1') #CAlibrationFactor
fid.write('\n') #new line

# fid.write('DOI capability:    -1,-1,-1') #copying from the cdh from GATEROOT path
# fid.write('\n')

# fid.write('Isotope: unknown') #CAlibrationFactor
# fid.write('\n') #new line

#don't know if this is needed (seems to be in documentation for PET but not SPECT)
#fid.write('Attenuation correction flag: '+str(ac_flag)) #attenuation correction flag
#fid.write('\n') #new line

fid.close()
```

# Writing the cdF file

```python
#write binary data file
fid = open(binary, 'wb') #open binary data file for writing

dtype1 = np.dtype(np.uint32).newbyteorder("L") #write t to file; 32 bit unsigned integer, little-endian
dtype2 = np.dtype(np.float32).newbyteorder("L") #write p to file; 32 bit float, little-endian

X,Y,Z = input_data_EM.shape[1], input_data_EM.shape[2], input_data_EM.shape[0]

for column in range(X):
    for row in range(Y):
        for projection in range(int(Z)): #loop through data values
            t = projection*proj_time #time stamp in ms (first projection at zero time)
            fid.write(np.array(t,dtype=dtype1)) #write t to file; 32 bit unsigned integer, little-endian
            p = input_data_EM[projection,row,column] #pixel data
            fid.write(np.array(p,dtype=dtype2)) #write p to file; 32 bit float, little-endian
            
            if scat_flag == 1:
                s = input_data_SCAT[row,column,projection] #scatter estimate
                fid.write(np.array(s,dtype2)) #write s to file; 32-bit float, little-endian
            if n == 1:
                n = 1.0 #normalisation value (could use to correct for uniformity but set to 1 for all pixels for now)
                fid.write(np.array(n,dtype=dtype2)) #write n to file; 32-bit float, little-endian
            
            v = projection #the projection ID (starts at 0)
            fid.write(np.array(v,dtype = dtype1)) #write v to file; 32 bit unsigned integer, little-endian
            
            c = column + row*X  #calculate the bin ID
            fid.write(np.array(c,dtype = dtype1)) #write c (bin ID) to file; 32 bit unsigned integer, little-endian

fid.close()
```
